```python
import json
import pandas as pd

```

```python
data = pd.read_csv("./voters_list.csv")
# data = data[:5]
data = data[1:]
result_dict = dict()
for row in data.values:
    try:
        row[1] = int(row[1])
        row[3] = int(row[3])
    except: ...
    cnic = str(row[-1])
    cnic = cnic.replace("-", "")
    result_dict[cnic] = list(row[:-1])
    # print(row)

file_path = "voters_list.json"

# Open the file in write mode and write the JSON data
with open(file_path, "w") as json_file:
    json.dump(result_dict, json_file)
```
